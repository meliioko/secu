```python
df = pd.read_excel('SWaT.A3_dataset_Jul 19_labelled.xlsx', skiprows=1)
df.drop(0, inplace=True)
df.head()
```

```python
print(df.Label.value_counts())
print(df.Attack.value_counts())
```

```python
df['Label'] = df['Label'].fillna(0)
df['Attack'] = df['Attack'].fillna('benign')
```

```python
print(df.Label.value_counts())
print(df.Attack.value_counts())
```

```python
columns = df.columns
columns = columns[2:]
for c in columns:
    df[c] = pd.to_numeric(df[c], errors='coerce')

df['GMT +0'] = pd.to_datetime(df['GMT +0'],  format='ISO8601')
```

```python
nan_rate_per_column = df.isnull().mean() * 100
broken_columns = []
print("Rate of NaN values for each column:")
for column, nan_rate in nan_rate_per_column.sort_values(ascending=False).items():
    print(f"{column}: {nan_rate:.2f}%")
    if nan_rate == 100:
        broken_columns.append(column)
```

```python
print(broken_columns)
```

```python
df.drop(broken_columns, axis=1, inplace=True)
```

```python
df.drop(['GMT +0', 'Attack'], axis=1, inplace=True)
```

```python
from sklearn.ensemble import IsolationForest

outliers_fraction = 0.03
clf = IsolationForest(contamination=outliers_fraction, random_state=42)
clf.fit(df)
outliers = clf.predict(df)
df['outliers'] = outliers
num_outliers = (df['outliers'] == -1).sum()
num_fraudulent_outliers = df[(df['outliers'] == -1) & (df['Label'] == 1)].shape[0]
# Print the number of outliers and attack outliers
print("Number of outliers detected:", num_outliers)
print("Number of outliers that are labeled as attack:", num_fraudulent_outliers)
# Print fraudulent outliers
fraudulent_outliers = df[(df['outliers'] == -1) & (df['Label'] == 1)]
print("Fraudulent outliers:")
print(fraudulent_outliers)
```

```python
from sklearn.neighbors import LocalOutlierFactor
outliers_fraction = 0.03
clf_lof = LocalOutlierFactor(contamination=outliers_fraction)
outliers_lof = clf_lof.fit_predict(df)
df['lof_outliers'] = outliers_lof
num_outliers_lof = (df['lof_outliers'] == -1).sum()
num_fraudulent_outliers_lof = df[(df['lof_outliers'] == -1) & (df['Label'] ==1)].shape[0]
print("Number of outliers detected by LocalOutlierFactor:", num_outliers_lof)
print("Number of outliers detected by LocalOutlierFactor that are labeled as␣fraudulent transactions:", num_fraudulent_outliers_lof)
```

```python
common_outliers = df[(df['outliers'] == -1) & (df['lof_outliers'] == -1)]
num_common_outliers = common_outliers.shape[0]
print("Number of common outliers between IsolationForest and LOF:", num_common_outliers)
```
